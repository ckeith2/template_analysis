```python
#switch to dnde
#find index in energy list closest to this energy
def find_nearest(energiesforBH, central_energies, lum):
    array = np.array(energiesforBH[::-1])
    backwardslum = np.array(lum[::-1])
    plt.plot(energiesforBH, lum*np.array(energiesforBH)**2)
    idxs = len(energiesforBH)-np.where(backwardslum*np.array(array)**2 >= np.nanmax(backwardslum)*np.array(array)**2/1e2)[0]
    value = energiesforBH[idxs[0]]
    
    array = np.asarray(central_energies)
    idx = (np.abs(array - value)).argmin()
    
    if idx < 2:
        return 2
    else:
        return idx

index_cut = find_nearest(energiesforBH, central_energies1, lum)
print(index_cut)
plt.close()
width = 4
fig = plt.figure(figsize = (12, 10))
plt.plot(energiesforBH, lum3*energiesforBH**2, color = lover_colors[-1], linewidth = width, label = '3e16 g', ls = '--')
plt.plot(energiesforBH, lum2*energiesforBH**2, color = lover_colors[-2], linewidth = width, label = '3e17 g', ls = ':')
plt.plot(energiesforBH, lum*energiesforBH**2, color = lover_colors[-3], linewidth = width, label = '3e18 g', ls = '-.')
plt.yscale('log')
plt.xscale('log')
plt.ylim(1e12, 1e23)
plt.xlim(1e-1, 1e4)
plt.ylabel(r'$\frac{dN_{\gamma}}{dE}E^{2}$')
plt.xlabel('$E$ (MeV)')
plt.legend()
plt.savefig('./paper_images/blackholespectra.pdf')
#plt.axvline(x=central_energies1[index_cut])
#plt.axvline(x=.1, color = 'black')
```
